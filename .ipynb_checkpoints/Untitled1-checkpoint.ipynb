```python
df.index
```

Index is a label not a row number

If we use the drop() func, it doesn't reset the index unless mentioned


```python
type(df.index)
```

This tells that this index is of type RangeIndex (where index is sequential, starts from 0 and increases step wise)

Pandas has multiple types of Indexes (int64, float64, datetime)

```python
df.loc[5]
```

loc is label based, it cares about the index.. It gave us the row whose index is 5

```python
df.iloc[5]
```

iloc doesnt care about index labels, it is postition based. It gave us the row whose position is 5
Coincidentally in this case both index 5 and position 5 lie at the same spot so both the commands have the same row
iloc is unaffected by index changes

```python
df_new = df.drop(5)
```

Doesn't modify/drop in the original DataFrame, if we had done df = df.drop(5) then it would have made the change so it is 
not in-place by default
drop command is label/index based

```python
df_new.loc[6]
```

```python
df_new.iloc[5]
```

By using the drop command we and then using loc and iloc we can now understand that loc is index based and iloc is position based

```python
df_new.loc[5]
```

```python
df.loc[0:5]
```

```python
df.iloc[0:5]
```
