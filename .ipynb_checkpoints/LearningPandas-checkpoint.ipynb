Weighted rate answers "How much money do we lose?"
Whereas Simple rate answers, How often do customers default?

Organizations care about money not just frequency.

Suppose total loan exposure in a segment is: 10,000,000
Weighted default we got say = 0.18

Then, 0.18 * 10,000,000 = 1,80,000 is the money at risk.
Mutiplying the fraction by total exposure gives us the money.

Suppose:

Segment A:

₹10M exposure

10% loss = ₹1M loss

18% interest income = ₹1.8M income
→ Net = +₹0.8M

Segment B:

₹1M exposure

20% loss = ₹0.2M loss

12% interest income = ₹0.12M income
→ Net = -₹0.08M

Now Segment B is actually worse.

This Is The Real Banking Equation

Banks don’t look at default rate alone.

They look at:

Expected Loss=PD×LGD×EAD

Where:

PD = Probability of Default

LGD = Loss Given Default

EAD = Exposure at Default

And compare that against:

Expected Revenue

```python
seg = df.groupby('savings_amount', observed=True).apply(
    lambda x: (x['credit_amount'] * (1 - x['default_risk'])).sum() / x['credit_amount'].sum(),
    include_groups=False)
```

```python
seg
```

.apply() lets you use multiple columns together.

-But it is slower as it loops over groups.
-Executes the lambda per group
-Not fully optimized in C


# Lets see an alternate way

```python
seg = (
    df.assign(default_flag = 1 - df['default_risk'])
    .assign(weighted_default = df['credit_amount'] * (1 - df['default_risk']))
    .groupby('purpose', observed=True)
    .agg(
        exposure=('credit_amount', 'sum'),
        default_exposure=('weighted_default', 'sum')
    )
)

seg['exposure_weighted_default_rate'] = seg['default_exposure'] / seg['exposure']
```

# Time Series & Window Functions

```python
#Creating a Time Dataset

payments = pd.DataFrame({
    'loan_id' : [101,101,101,102,102,103,103,103],
    'date' : [
        '2023-01-10', '2023-02-10', '2023-03-10',
        '2023-01-15', '2023-02-15',
        '2023-01-05', '2023-02-05', '2023-03-05'
    ],
    'payment_amount' : [1000,1200,1100,1500,1500,800,900,1000]
})

```

```python
payments
```

```python
payments['date'] = pd.to_datetime(payments['date'])
```

The data column was a string data type column, changed that into datetime

```python
payments.dtypes
```

```python
#Extracting Date Components

payments['year'] = payments['date'].dt.year
payments['month'] = payments['date'].dt.month
payments['day'] = payments['date'].dt.day
```

```python
payments
```

This is useful for monthly or yearly analysis.

```python
#Sorting by Time
#Time operations require proper ordering


payments = payments.sort_values(['loan_id', 'date'])
```

```python
payments
```

```python
#Rolling Window (Moving Average) --> average of last 2 payments

payments['rolling_avg'] = (payments.groupby('loan_id')['payment_amount']
                           .rolling(2)
                           .mean()
                           .reset_index(level=0, drop=True)
                          )
                        
```

```python
payments
```

```python
#Expanding Window (Cumulative Average)

payments['expanding_avg'] = (payments.groupby('loan_id')['payment_amount']
                             .expanding()
                             .mean()
                             .reset_index(level=0, drop=True)
                            )
```

```python
payments
```

```python
payments['cumulative_paid'] = (payments.groupby('loan_id')['payment_amount']
                               .cumsum()
                              )
```

```python
payments
```

```python
#Lag Feature (shows the previous payment values. Heavily used in credit risk modeling)

payments['prev_payment'] = (payments.groupby('loan_id')['payment_amount']
                            .shift(1)
                           )
```

```python
payments
```

This set of functions that we saw in Time Series Functions allows us to analyze:

    1. repayment trend
    2. spending patterns
    3. delinquency detection
    4. rolling default metrics
    5. maintains time logic

* Before using operations like:
    1. rolling()
    2. shift()
    3. expanding()
    4. cumsum()

You should ensure the data is sorted by time. Other wise the calculation follow the row order instead of time order.

```python
#Time Based Grouping

payments.groupby('month')['payment_amount'].sum()
```

```python
payments.groupby('year')['payment_amount'].sum()
```

```python
#Resampling : Instead of extracting a column, pandas automatically group time intervals

payments = payments.set_index('date')
```

```python
payments.resample('M')['payment_amount'].sum()
```

```python
payments.resample('W')['payment_amount'].sum()
```

Banks often track:
    - monthly loan repayments
    - weekly transaction volume
    - quaterly default rate
